# **조건에 맞는 데이터 추출**

```python
import pandas as pd

df = pd.DataFrame({'a' : [i for i in range(1, 11)], 'b' : [i for i in range(11, 21)], 'c' : [i for i in range(21, 31)]})
df
```

a, c 열 출력

```python
df[['a', 'c']]
```

a 가 3 이상인 데이터 출력

```python
df[df['a'] >= 3]
```

a 가 3 이상인 데이터 중 a, c 열만 출력

```python
df[df['a'] >= 3][['a', 'c']]
```

a 가 3 이상이고, b 가 16 미만인 데이터 출력

```python
df[(df['a'] >= 3) & (df['b'] < 16)]
```

```python
a = (df['a'] >= 3) & (df['b'] < 16)
a
```

```python
type(a)
```

```python
df[a]
```

a 가 3 이하 이거나 7 이상인 데이터를 출력

```python
df[(df['a'] <= 3) | (df['a'] >= 7)]
```

a 가 3 이상이고, b 가 16 미만이거나 c 가 30 인 데이터를 출력

```python
df[(df['a'] >= 3) & ((df['b'] < 16) | (df['c'] == 30))]
```

*   결측값을 다른 값으로 대체

```python
df = pd.DataFrame({'a' : [1, 1, 3, 4, 5], 'b' : [2, 3, np.nan, 3, 4], 'c' : [3, 4, 7, 6, 4]})
df
```

-특정값으로 대체

```python
df.fillna(0, inplace=True)
df
```

- 앞이나 뒤의 숫자로 바꾸기

```python
df = pd.DataFrame({'a' : [1, 1, 3, 4, np.nan], 'b' : [2, 3, np.nan, np.nan, 4], 'c' : [np.nan, 4, 1, 1, 4]})
df
```

1) 뒤의 값으로 채우기

```python
df.fillna(method='bfill')
```

2) 앞의 값으로 채우기

```python
df.fillna(method='ffill')
```

- limit 설정

```python
df.fillna(method='ffill', limit =1)
```

데이터 프레임에 존재하는 결측값들을 뒤의 값으로 대체한 이후 앞의 값으로 대체하시오

```python
df = pd.DataFrame({'a' : [1, 1, 3, 4, np.nan], 'b' : [2, 3, np.nan, np.nan, 4], 'c' : [np.nan, 4, 1, 1, 4]})
df
```

```python
df.fillna(method='bfill', inplace=True)
df
```

```python
df.fillna(method='ffill', inplace=True)
df
```

- 평균으로 대체

```python
df = pd.DataFrame({'a' : [1, 1, 3, 4, np.nan], 'b' : [2, 3, np.nan, np.nan, 4], 'c' : [np.nan, 4, 1, 1, 4]})
df
```

```python
df.mean()['a']
```

```python
df.fillna(df.mean()['a'])
```

```python
df.mean()
```

```python
df.fillna(df.mean())
```

b, c 의 결측값들을 데이터 프레임의 전체 값의 평균으로 치환

```python
df.fillna(df.mean()[['b','c']])
```
